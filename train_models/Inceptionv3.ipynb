```python
import matplotlib.pyplot as plotter_lib

import numpy as np

import PIL as image_lib

import tensorflow as tflow

from tensorflow.keras.layers import Flatten

from keras.layers.core import Dense

from tensorflow.keras.models import Sequential

from tensorflow.keras.optimizers import Adam

import matplotlib as plt

import pickle

from keras.models import Sequential
from keras.models import Model
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping, ReduceLROnPlateau, TensorBoard
from keras import optimizers, losses, activations, models
from keras.layers import Convolution2D, Dense, Input, Flatten, Dropout, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D, Concatenate
from keras import applications
```

```python
data_directory="/Users/madalena/Documents/projeto2/train"
```

```python
img_height,img_width=180,180

batch_size=32
train_ds,validation_ds=tflow.keras.utils.image_dataset_from_directory(
    data_directory,
    labels='inferred',
    label_mode='categorical',
    color_mode='rgb',
    batch_size=32,
    image_size=(img_height, img_width),
    shuffle=True,
    validation_split=0.2,
    subset='both',
    seed=123
)
```

# Regular

```python
modelv3=Sequential()
batch_size = 16

basev3=tflow.keras.applications.InceptionV3(
    include_top=False,
    weights='imagenet',
    input_shape=(180,180,3),
    pooling='avg',
    classes=85
)

epochs=200

basev3.trainable=True
modelv3.add(basev3)

#softmax
modelv3.add(Dense(85, activation='softmax'))

modelv3.compile(loss='categorical_crossentropy', 
              optimizer=optimizers.SGD(lr=1e-4, 
                                       momentum=0.9),
              metrics=['accuracy'])
modelv3.summary()

history = modelv3.fit(train_ds,validation_data=validation_ds,epochs=epochs,
                             
                              shuffle=True, 
                              verbose=True,
                              #callbacks=[early]
                              )
modelv3.save("inceptionv3")

with open("/Users/madalena/Documents/projeto2/train_models/iv3", 'wb') as file_pi:
    pickle.dump(history.history, file_pi)
```

# Transfer

```python
tmodelv3=Sequential()
batch_size = 16

basev3=tflow.keras.applications.InceptionV3(
    include_top=False,
    weights='imagenet',
    input_shape=(180,180,3),
    pooling='avg',
    classes=85
)

epochs=200

basev3.trainable=False
tmodelv3.add(basev3)

#softmax
tmodelv3.add(Dense(85, activation='softmax'))

tmodelv3.compile(loss='categorical_crossentropy', 
              optimizer=optimizers.SGD(lr=1e-4, 
                                       momentum=0.9),
              metrics=['accuracy'])
tmodelv3.summary()

thistory = tmodelv3.fit(train_ds,validation_data=validation_ds,epochs=epochs,
                             
                              shuffle=True, 
                              verbose=True,
                              #callbacks=[early]
                              )
tmodelv3.save("inceptionv3transfer")

with open("/Users/madalena/Documents/projeto2/train_models/iv3t", 'wb') as file_pi:
    pickle.dump(thistory.history, file_pi)
```
